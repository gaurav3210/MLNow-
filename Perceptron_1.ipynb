# Problem Statement
Train a model to learn  the table of 10
Equation "y = 10*x"


##Data Creation##

x = [0 1 2 3 4 5 .. ]

y = [0 10 20 30 40 50...]

```python
x = [i for i in range(21)];  #list comprehension
print(x)
```

```python
y = [i for i in range(20*10+1) if i%10==0]
print(y)
```

##Approach 1## 
Defining a function 'y=10x'

```python
def tempFunc(x):
  y=10*x
  return y;

for value in x:
  print(tempFunc(value))
```

##Approach 2##
I'm super enthusiastic about ML so I'm gonna make a redundant ML model for the table of 10
you going to need xTrain, yTrain, xTest and yTest from the x and y lists 

```python
print(f'this is x: {x}');
print(f'this is y: {y}');
```

```python
xTrain = x[:-5] ##training data
yTrain = y[:-5] ##training label

xTest = x[-5:] ## test data
yTest = y[-5:] ## test label

print(f'''Training Data: 
xTrain: {xTrain}
yTrain: {yTrain}

Testing Data:
xTest: {xTest}
yTest: {yTest}
''')
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
##perceptron model
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(units=1, input_shape=[1])  ##units is the number of neurons and input_shape is shape of input of your neuron
])
```

```python
model.compile(optimizer='adam',loss='mae')  #sgd, rmsprop, adamax,adagrad; loss--> mse,mae (this is regression example)
```

```python
model.fit(x=xTrain,y=yTrain,validation_data=(xTest,yTest), epochs=10000);
```
